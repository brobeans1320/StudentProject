# [dbo].[createStudent]

```sql
/****** Object:  StoredProcedure [dbo].[createStudent]    Script Date: 6/7/2024 10:04:48 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[createStudent]
    @firstName VARCHAR(25),
    @lastName VARCHAR(25),
    @dateOfBirth DATE,
    @age INT,
    @major VARCHAR(30),
    @gpa FLOAT
AS
BEGIN
    INSERT INTO students (firstName, lastName, dateOfBirth, age, major, gpa)
    VALUES (@firstName, @lastName, @dateOfBirth, @age, @major, @gpa);
END;
GO
```

# [dbo].[deleteStudent]

```sql
/****** Object:  StoredProcedure [dbo].[deleteStudent]    Script Date: 6/7/2024 10:04:48 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[deleteStudent]
    @id INT
AS
BEGIN
    DELETE FROM students
    WHERE id = @id;
END;
GO
```

# [dbo].[selectAllStudents]

```sql
/****** Object:  StoredProcedure [dbo].[selectAllStudents]    Script Date: 6/7/2024 10:04:48 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[selectAllStudents]
AS
BEGIN
    SELECT 
        id,
        firstName,
        lastName,
        dateOfBirth,
        age,
        major,
        gpa
    FROM students;
END;
GO
```

# [dbo].[updateStudent]

```sql
/****** Object:  StoredProcedure [dbo].[updateStudent]    Script Date: 6/7/2024 10:04:48 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[updateStudent]
    @id INT,
    @firstName VARCHAR(25),
    @lastName VARCHAR(25),
    @dateOfBirth DATE,
    @age INT,
    @major VARCHAR(30),
    @gpa FLOAT
AS
BEGIN
    UPDATE students
    SET
        firstName = @firstName,
        lastName = @lastName,
        dateOfBirth = @dateOfBirth,
        age = @age,
        major = @major,
        gpa = @gpa
    WHERE id = @id;
END;
GO
```
